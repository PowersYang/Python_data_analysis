```python
import matplotlib.pyplot as plt
```

### 传数据的两种方式

```python
fig, ax = plt.subplots()

ax.plot([2, 2, 2, 2])

plt.show()
```

```python
fig, ax = plt.subplots()

data = {'a': [1, 2, 3], 'b': [3, 2, 1]}
ax.plot('a', 'b', data=data)

plt.show()
```

### 绘制多条线

```python
fig, ax = plt.subplots()

data = {'a': [1, 2, 3], 'b': [3, 2, 1]}

ax.plot('a', 'b', data=data)
ax.plot([2, 2, 2, 2])

plt.show()
```

### 其它设置

#### 1、颜色 —— color

可选值：
``'b'``          blue
``'g'``          green
``'r'``          red
``'c'``          cyan
``'m'``          magenta
``'y'``          yellow
``'k'``          black
``'w'``          white

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5, 6], color='r', drawstyle='steps')

plt.show()
```

#### 2、线性 —— linestyle

可选值：['solid' | 'dashed', 'dashdot', 'dotted' | (offset, on-off-dash-seq) | ``'-'`` | ``'--'`` | ``'-.'`` | ``':'`` | ``'None'`` | ``' '`` | ``''``]

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5, 6], color='r', linestyle='dotted')

plt.show()
```

#### 3、drawstyle

可选值：['default' | 'steps' | 'steps-pre' | 'steps-mid' | 'steps-post'] 

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5, 6], color='r', drawstyle='steps')

plt.show()
```

#### 4、标记 —— marker

可选值：
``'.'``          point marker
``','``          pixel marker
``'o'``          circle marker
``'v'``          triangle_down marker
``'^'``          triangle_up marker
``'<'``          triangle_left marker
``'>'``          triangle_right marker
``'1'``          tri_down marker
``'2'``          tri_up marker
``'3'``          tri_left marker
``'4'``          tri_right marker
``'s'``          square marker
``'p'``          pentagon marker
``'*'``          star marker
``'h'``          hexagon1 marker
``'H'``          hexagon2 marker
``'+'``          plus marker
``'x'``          x marker
``'D'``          diamond marker
``'d'``          thin_diamond marker
``'|'``          vline marker
``'_'``          hline marker

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5, 6], marker='o')

plt.show()
```
