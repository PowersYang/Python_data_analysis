```python
import matplotlib.pyplot as plt
```

### 传数据的两种方式

```python
fig, ax = plt.subplots()

ax.plot([2, 2, 2, 2])

plt.show()
```

```python
fig, ax = plt.subplots()

data = {'a': [1, 2, 3], 'b': [3, 2, 1]}
ax.plot('a', 'b', data=data)

plt.show()
```

### 绘制多条线

```python
fig, ax = plt.subplots()

data = {'a': [1, 2, 3], 'b': [3, 2, 1]}
ax.plot('a', 'b', data=data)

ax.plot([2, 2, 2, 2])

plt.show()
```

### 其它设置

```python
fig, ax = plt.subplots()

plt.plot([1, 2, 3, 4, 5, 6], color='r', marker='+', linestyle='--', linewidth='2')

plt.show()
```
