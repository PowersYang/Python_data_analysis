Matplotlib是一个Python 2D绘图库，可以生成各种硬拷贝格式和跨平台交互式环境的出版物质量数据。Matplotlib可用于Python脚本，Python和IPython shell，Jupyter笔记本，Web应用程序服务器和四个图形用户界面工具包。

Matplotlib试图让简单易事的事情成为可能。你只需几行代码即可生成绘图，直方图，功率谱，条形图，误差图，散点图等。

### **简单示例**

```python
import matplotlib.pyplot as plt

# 创建数据集
x = np.linspace(0, 2, 100)

# 绘制三条线
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# 设置x轴和y轴标记
plt.xlabel('x label')
plt.ylabel('y label')

# 设置title
plt.title("Simple Plot")

# 显示legend
plt.legend()
plt.show()
```

### **理解Figure和Axes**

- **Figure**：整个画布对象。我们所有的绘图操作都是在Figure对象上操作的，不论是单个图表还是多子图。

- **Axes**：这就是我们平常说的具体的“一幅图”（），它是具有数据空间的图像区域。
